- 입력1: [1, 2, 3, 4] 출력1: [3, 4, 5, 6, 7] 
- 설명1: 가능한 모든 합은 3 (1+2), 4 (1+3), 5 (1+4), 5 (2+3), 6 (2+4), 7 (3+4)
- 여기서 고유한 값들만 추려내고 오름차순으로 정렬합니다. 
- 입력2: [5, -2, 4] 출력2: [2, 3, 9]

- 고유한 값들을 오름차순으로 정렬합니다.

```python
# 숫자 입력
L = list(map(int, input("숫자들을 ',' 구분하여 입력하세요 : ").split(',')))

# 입력된 수의 가능한 합 찾기
def find_unique_sums(numbers:list[int]) -> list :
    # 결과를 저장할 리스트
    result = []

    # 리스트 길이 만큼 반복
    for i in range(len(numbers)) :
        # 뒤에 숫자와 비교해야하기 때문에 i+1
        for j in range(i+1, len(numbers)) :
            # 결과 리스트에 합 저장
            result.append(numbers[i]+ numbers[j])

    # 결과 리스트 정렬
    result = sorted(result)

    # 결과 반환
    return result

# 결과 출력
print(find_unique_sums(L))
```
