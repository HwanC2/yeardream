<aside>
수업 시간에 배웠던 Stack 구조를 Class로 구현해보세요.

**Class에는 다음 함수와 기능이 포함되어야 합니다.**

> push, pop, isFull, isEmpty, findTop

</aside>

```python
class Stack :
    # 생성자
    def __init__(self, max_size) :
        # 스택을 저장할 리스트와 스택의 최대 크기 설정
        self.stack = []
        self.max_size = max_size

    # 스택 추가 메서드
    def push(self, item) :
        # 스택이 가득 차지 않았다면, 스택 추가
        if not self.isFull() :
            print("스택 추가 : ", item)
            self.stack.append(item)
        else :
            print("스택이 가득 찼다!")

    # 스택 제거 메서드
    def pop(self) :
        # 스택이 비어있지 않으면,
        if not self.isEmpty() :
            # 스택의 맨 위 항목을 제거하고 반환.
            del self.stack[self.findTop()]
        else :
            print("스택이 비었다!")

    # 스택이 가득 찼는지 여부 판단 메서드
    def isFull(self) :
        return len(self.stack) == self.max_size
    
    # 스택이 비었는지 여부 판단 메서드
    def isEmpty(self) :
        return len(self.stack) == 0

    # 스택이 비어있지 않으면, 스택의 맨 위 항목을 제거하고 반환하는 메서드
    def findTop(self) :
        if not self.isEmpty() :
            print("스택 제거 : ", self.stack[-1])
            return -1
        else :
            return 0
            
        
stack = Stack(5)    # 스택 클래스 변수 생성

stack.push(1)       # 스택 추가
stack.push(2)       # 스택 추가
stack.push(3)       # 스택 추가
stack.push(4)       # 스택 추가
stack.push(5)       # 스택 추가
stack.push(6)       # 스택 추가시 가득참 에러 메세지 출력

stack.pop()         # 스택 제거
stack.pop()         # 스택 제거
stack.pop()         # 스택 제거
stack.pop()         # 스택 제거
stack.pop()         # 스택 제거
stack.pop()         # 스택 제거시 비었음 에러 메세지 출력
```

```python
# 중간값부터 비교하는 이진탐색 메서드
def binary_search(list:list, min, max, key):
    # 이진 탐색을 위해 최소값과 최대값 설정
    while min < max :
        # 중간값 찾기
        mid = (min + max) // 2

        # 만약 중간값이 찾고자 하는 값보다 작다면, 최소값을 중간값보다 하나 큰 값으로 설정
        if list[mid] < key :
            min = mid + 1
        # 만약 중간값이 찾고자 하는 값보다 크다면, 최대값을 중간값으로 설정
        elif list[mid] > key :
            max = mid
        # 만약 중간값이 찾고자 하는 값과 같다면, 중간값을 반환
            return mid

    # 리스트에 찾는 값이 없을 경우 -1 반환
    return -1
            

L = [1, 3, 5, 7, 9]     # 리스트 생성
x = 7                   # key 값으로 사용할 변수 생성

# 리스트에 포함된 값이 있는지 검색 후 결과 저장
result = binary_search(L, 0, len(L), x)

# 검색 결과 확인
if result == -1:
    print("찾는 값이 리스트에 없습니다.")
else:
    print(f"{x}은(는) 리스트의 인덱스 {result1}에 위치합니다.")
```

```python
# 은행 계좌 클래스
class BankAccount :
    # 생성자
    def __init__(self, account_number, owner) :
        self.account_number = account_number    # 계좌 번호
        self.owner = owner                      # 소유자
        self.balance = 0.0                      # 계좌 잔액
        self.transactions = []                  # 거래

        print("계좌 개설 완료.")    # 계좌 생성 완료 메세지 출력
        self.display_balance()     # 계좌 잔액 출력

    # 입금 메서드
    def deposit(self, amount) :
        self.balance += amount
        self.transactions.append({'type':'diposit','amount':amount, 'balance':self.balance})
        #print(f"{amount}원 입금 완료.")
        #self.display_balance()

    # 출금 메서드
    def withdraw(self, amount) :
        if self.balance >= amount :
            self.balance -= amount
            self.transactions.append({'type':'withdraw','amount':amount, 'balance':self.balance})
            #print(f"{amount}원 출금 완료.")
        else :
            print("계좌 잔액이 부족합니다.")
        #self.display_balance()
        
    # 계좌 잔액 출력 메서드
    def display_balance(self) :
        print(f"현재 잔액 : {self.balance}원")

    # 거래 내역 출력
    def print_transactions(self) :
        print("Recent transactions (most recent first) : ")
        # 리스트 역방향 출력
        for transaction in self.transactions[::-1] :
            print(transaction)


tossbank = BankAccount(140916, '콩지')  # 계좌 개설
tossbank.deposit(1000)                  # 입금
tossbank.withdraw(300)                  # 출금
tossbank.deposit(700)                   # 입금
tossbank.deposit(550)                   # 입금
tossbank.deposit(27000)                 # 입금
tossbank.withdraw(5800)                 # 출금
tossbank.withdraw(6000)                 # 출금
tossbank.withdraw(3600)                 # 출금
tossbank.withdraw(50000)                # 출금
tossbank.print_transactions()           # 거래 내역 조회
```
