## 문제 1: 비밀번호 강도 체크
사용자로부터 비밀번호를 입력받아, 그 비밀번호가 강한지 약한지를 평가하는 프로그램을 작성하세요. 비밀번호 강도는 다음 조건에 따라 결정됩니다:

8자리 이상, 숫자와 영문자(대문자 및 소문자)를 모두 포함: "강함"
그렇지 않으면: "약함"

[HINT] str.isalpha(), str.isdigit(), str.isalnum()

```python
password = "abnaklsfjklas1234"


# 1)
if len(password) >= 8:
    if password.isalnum():
        if password.isalpha(): # aaaaaaaaaa
            print("약함")
        elif password.isdigit(): # 111111111
            print("약함")
        else:  # a1b2c3d4d5
            print("강함")
    else:    # a1b2c3d4**
        print("약함")

else: # aaaaa
    print("약함")


# 2)
if len(password) >= 8 and password.isalnum():
        if password.isalpha(): # aaaaaaaaaa
            print("약함")
        elif password.isdigit(): # 111111111
            print("약함")
        else:  # a1b2c3d4d5
            print("강함")
else: # aaaaa, a1b2c3d4**
    print("약함")
```

## 문제 2. 별 찍기!

* 하늘의 아름다운 별을 따다 코드로 작성해보자.

> 이해하고 싶으신 분들은 별 찍기의 출력을 반대로 해보세요!

```python
# number = 7
*******
******
*****
****
***
**
*
```

```python
number = int(input("Input the number of lines : "))
# 1)
for line in range(number, 0, -1):
    for i in range(line): # 0, 1, 2, ..., number-1
        print('*', end="")
    print()

# 2)
for line in range(number):
    # number, number-1, ..., 4, 3, 2, 1 --- a
    # 0, 1, 2, 3, 4, 5, number-1 --- b
    # a + b = number <=> a = number-b
    for i in range(number-line):
        print('*', end='')
    print()

# 3)
for line in range(number, 0, -1):
    print('*' * line)
```
