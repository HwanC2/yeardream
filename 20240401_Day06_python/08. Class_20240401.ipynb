- 실재하는 무언가의 "개념"을 코드로 만들 때, 데이터 & 기능을 함께 구현할 수 있게 하려면, Class단위로 만드는게 좋다!

- Notebook이라는 개념을 클래스로 구현하였습니다.

- Notebook에서 필요한 여러가지 기능을 함수(class method)로 구현합니다.

```python
# 클래스 내부의 함수를 구현합니다.
# 해당 노트북의 운영체제가 UNIX 계열 운영체제인지 아닌지 확인하는 Class Method를 구현합니다.

class Notebook:
  ## self에는 이 함수를 호출한 객체의 이름이 들어갑니다.
  def __init__(self, manufacturer, model, ram_size, ssd_size, os):
    self.manufacturer = manufacturer
    self.model = model
    self.ram_size = ram_size
    self.ssd_size = ssd_size
    self.os = os

  ## TO-DO ##
  def is_Unix(self):
   '''
   OS가 윈도우 계열이면 UNIX 계열이 아니고, Linux나 macOS면 UNIX 계열이다.
   '''
   if self.os == 'windows':
     return False
   elif self.os == 'macOS':
     return True
   elif self.os == 'linux':
     return True
   
# 객체(object) <-> 클래스(class)
nb1 = Notebook(manufacturer='Samsung',
               model='Galaxy Book',
               ram_size=32,
               ssd_size=1024,
               os='windows')

nb2 = Notebook('Apple', 'Macbook Pro', 32, 1024, os='macOS')
nb3 = Notebook('Dell', 'XPS 9500', 16, 2048, 'linux')

print(nb1.os, nb2.os, nb3.os)
print(nb1.is_Unix(), nb2.is_Unix(), nb3.is_Unix())
```

```python
# 해당 노트북의 모델을 출력해주는 Class Method를 구현합니다.
# 클래스 내부의 함수를 구현합니다.
# 해당 노트북의 운영체제가 UNIX 계열 운영체제인지 아닌지 확인하는 Class Method를 구현합니다.

class Notebook:
  ## self에는 이 함수를 호출한 객체의 이름이 들어갑니다.
  def __init__(self, manufacturer, model, ram_size, ssd_size, os):
    self.manufacturer = manufacturer
    self.model = model
    self.ram_size = ram_size
    self.ssd_size = ssd_size
    self.os = os

  ## TO-DO ##
  def is_Unix(self):
   '''
   OS가 윈도우 계열이면 UNIX 계열이 아니고, Linux나 macOS면 UNIX 계열이다.
   '''
   if self.os == 'windows':
     return False
   elif self.os == 'macOS':
     return True
   elif self.os == 'linux':
     return True
   
   ## TO-DO ##
  def print_model(self):
    print(f"Model : {self.model}")
   
# 객체(object) <-> 클래스(class)
nb1 = Notebook(manufacturer='Samsung',
               model='Galaxy Book',
               ram_size=32,
               ssd_size=1024,
               os='windows')

nb2 = Notebook('Apple', 'Macbook Pro', 32, 1024, os='macOS')
nb3 = Notebook('Dell', 'XPS 9500', 16, 2048, 'linux')

nb1.print_model()
nb2.print_model()
nb3.print_model()
```

#### 클래스 상속 (Class inheritence)

- Notebook들은 제조사마다 다른 모델이 있습니다.


- 해당 모델들은 Notebook이지만, 모델마다 다른 특징을 가지고 있습니다.


- 노트북의 개념을 그대로 이어받아, 모델마다 다른 개념을 Class로 구현해봅시다.

```python
# notebook class를 상속받아서 새로운 MacBook class를 정의
## 클래스 정의할 때 괄호안에 특정 클래스의 이름을 넣어주면 상속이 됩니다.
## Notebook(superclass) ---> MacBook(subclass)
class MacBook(Notebook):
    def __init__(self, manufacturer, model, ram_size, ssd_size, os):
        super().__init__(manufacturer, model, ram_size, ssd_size, os) # superclass의 생성자를 그대로 쓰겠다.

    def check_model(self):
        '''
        이 객체의 model이 맥북 라인업이 맞는지 체크하는 함수.
        '''
        if self.model in ['MacBook Pro', 'MacBook Air', 'MacBook']:
            return True
        else:
            return False
```

```python
# notebook class를 상속받아서 새로운 Dell_Laptop class를 정의
class DellLaptop(Notebook):
    def __init__(self, manufacturer, model, ram_size, ssd_size, os):
        super().__init__(manufacturer, model, ram_size, ssd_size, os)

    ## TO-DO ##
    def is_Dell(self):
        '''
        이 객체가 Dell 노트북이 맞는지 체크하는 함수.
        '''
        if self.manufacturer == 'Dell':
            return True
        else:
            return False
```

#### Method Overriding

```python
# 각 모델마다 다르게 정보를 출력해주는 is_UNIX 함수를 재정의합니다.
class MacBook(Notebook):
    def __init__(self, manufacturer, model, ram_size, ssd_size, os):
        super().__init__(manufacturer, model, ram_size, ssd_size, os) # superclass의 생성자를 그대로 쓰겠다.

    def is_Unix(self):
        ## Method Overriding ##
        if self.os in ['linux', 'Sonoma', 'Ventura', 'Catelina', 'Mojave']:
            return True
        else:
            return False
        
    def check_model(self):
        '''
        이 객체의 model이 맥북 라인업이 맞는지 체크하는 함수.
        '''
        if self.model in ['MacBook Pro', 'MacBook Air', 'MacBook']:
            return True
        else:
            return False
```

```python
mb = MacBook('Apple', 'MacBook Pro', 32, 1024, 'Sonoma')
#mb.check_model()
mb.is_Unix() ## superclass에 있는 is_Unix가 아니고, MacBook class에서 새롭게 정의한 is_Unix 함수가 됩니다.
```
